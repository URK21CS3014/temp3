```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_excel('/content/Titanic_visu (1).xlsx')
print(df)
```

```python
plt.figure(figsize=(12,5))
data=df['Embarked'].value_counts()
plt.bar(data.index,data.values,color=['yellow','orange','black'])
```

```python
import pandas as pd
import numpy as np
N = 15
plt.figure(figsize = (15, 15))
#see the spelling of arrange
ind = np.arange(N)
width = 0.25 
xval = df['Fare'].head(15)

bar1 = plt.bar(ind, xval, width, color = 'y')
yval = df['New_Fare'].head(15)
bar2 = plt.bar(ind + width, yval, width, color = 'b')

#comparitive charts should always have legends

plt.xlabel('Name')
plt.ylabel('Fare')
plt.xticks(ind + width, df['Name'].head(15))
plt.legend((bar1, bar2),('Fare', 'New_Fare'))
```

```python
plt.barh(df['Embarked'],df['Fare'] )
plt.title("Embarked vs Fare")
plt.xlabel("Embarked")
plt.ylabel("Fare")
```
